# OHBM Brainhack 2017
Takes dtiprep output report file (dwi_QCReport.txt) outputs images of gradients with bad slices and marks bad slices

```python
import re
from io import StringIO
import pandas as pd
from nilearn import plotting, image
import numpy as np

%matplotlib inline
```

```python
import numpy.linalg as npl
import nibabel as nib
def mm_2_vox(img, coords):
    #coords = (x,y,z)    
    return nib.affines.apply_affine(npl.inv(img.affine), coords)
def vox_2_mm(img, coords):
    return nib.affines.apply_affine(img.affine, coords)
```

```python

results_file = "/Users/franzliem/Desktop/DWI/dti.062317_155056/dwi_QCReport.txt"

with open(results_file) as fi:
  s = fi.read()

co = re.compile(r'Slice-wise Check Artifacts:[\s\n\t\w\d\#\.]*')
mo = co.search(s)
g = mo.group()
gg="\n".join(g.split("\n")[1:])

data = StringIO(gg)
df = pd.read_csv(data, sep="\t")


gradients_w_bad_slices = df.iloc[:,1].unique()
print(gradients_w_bad_slices)
```

```python
dwi_file = "/Users/franzliem/Desktop/DWI/data/sub-01/dwi/sub-01_dwi.nii.gz"
```

```python
for i in gradients_w_bad_slices:
    im = image.index_img(dwi_file, i)
    vmax = np.percentile(im.get_data().flatten(), 99)
    slices = df.loc[df.iloc[:,1]==i].iloc[:, 2].values
    print(slices)
    d = plotting.plot_anat(im, vmax=vmax, draw_cross=False, title="gradient %s; slices %s"%(i,slices), cut_coords=(0,0,0))
    
    for z_vox in slices:
        coords = vox_2_mm(im, (0,2,z_vox))
        d.add_markers(marker_coords=[( 0, 0,  coords[2])], marker_color='r', marker_size=10)
```
